## Problem: Scrape infobox from The Batman(wikipedia)

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
URL = requests.get('https://en.wikipedia.org/wiki/The_Batman_(film)')

#convert to beautifulsoup object
soup = bs(URL.content)
contents = soup.prettify()
print(contents)
```

```python
info_box = soup.find(class_ = 'infobox vevent')
info_rows = info_box.find_all("tr")
for row in info_rows:
    print (row.prettify())
```

```python
movie_info = {}


def get_content_value(row_data):
    if row.find("li"):
        return [li.get_text().replace("\xa0", " ") for li in row_data.find_all("li")]
    else:
        return row_data.get_text().replace("\xa0", " ")

# 'index' is the index of the current element in the iterable
# 'row' is the value of the current element in the iterable
for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['title'] = row.find("th").get_text(" ", strip = True)
    elif index == 1:
        continue
    else:
        content_key = row.find("th").get_text(" ", strip = True)
        content_value = get_content_value(row.find("td"))
        movie_info[content_key] = content_value

movie_info
```

## Problem: save and reload movie data
